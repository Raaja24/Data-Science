## 6) Plot

plot sctter plot between violent_crimes and rapes column.

found that where the crime rete is low there number of rape cases  is also low.

```python
sns.scatterplot(x= 'violent_crimes', y = 'rapes' ,data = crime_data);
```

Plot pie chart using matplotlib and plt. pie command to plot and use explode function to show a particular column.

found that assaults crime is around 49 % whereas the rapes crime rate is barely 4.63 %
found that the robberies case is near 44 % whereas the homicides rate is nearly 1.395 %.

```python
total_homicides_crime = crime_data.homicides.sum()
total_rapes_crime = crime_data.rapes.sum()
total_assaults_crime = crime_data.assaults.sum()
total_robberies_crime = crime_data.robberies.sum()

plt.figure(figsize=(6,6))

label = ['Homicides','Rapes','Assaults','Robberies']
myexplode = [0, 0.5, 0, 0]
plt.pie([total_homicides_crime,total_rapes_crime,total_assaults_crime,total_robberies_crime], labels=label, autopct='%.3f %%',explode = myexplode)
plt.title('Total number of crime divided into these categories')

plt.show()
```
